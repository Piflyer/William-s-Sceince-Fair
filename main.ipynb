# Sorting By Randomly obtaining 20 sites

```python
def get_random_sites(df, n):
    current_sites = df['lng'].unique()
    current_sites = list(current_sites)
    print('There are {} sites in the dataset'.format(len(current_sites)))
    df = df.drop_duplicates(subset='reference_no')
    print('There are {} unique sites in the dataset'.format(df.shape[0]))
    output = pd.DataFrame()
    output = pd.DataFrame()
    counter = 0
    ultra_state = True
    while ultra_state:
        state = True
        while state:
            if current_sites == []:
                print('No more sites to add')
                ultra_state = False
                break
            random_site = np.random.choice(current_sites)
            if df[df['lng']==random_site].shape[0] > 4:
                output = pd.concat([output, df[df['lng']==random_site]])
                print('Site #{} added'.format(counter+1))
                state = False
            current_sites.remove(random_site)
        counter += 1
        if counter == n:
            ultra_state = False
    return output
```

```python
random_sample = get_random_sites(loc, 20)
```
